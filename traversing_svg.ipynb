```python
import os
import xml.etree.ElementTree as ET
import pandas as pd
```

```python
svg_file = os.path.join( 'resources','gene_page_graphic.svg' )
tree = ET.parse( svg_file )
root = tree.getroot()
```

```python
csv_file = os.path.join( 'resources','10_human_brain_atlas_full_graph.csv')
structures = pd.read_csv(csv_file)
structures.set_index('id', inplace=True)
```

```python
print("Schematics: ")

for sch in root.findall('g') :
    
        print("--", sch.attrib['label'])
        
        for p in sch.findall('path') :
            
            if 'structure_id' in p.attrib.keys() :
                
                sid = int(p.attrib['structure_id'])
                print("---", structures.loc[sid,'acronym'], ":", structures.loc[sid,'name'], "(",sid,")" )
                
                ### you can modify the color of the polygon based on structure 
                ### by modifying the SVG attribute'style'
                
                
```
